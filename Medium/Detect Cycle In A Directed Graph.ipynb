# DETECT CYCLE IN A DIRECTED GRAPH

Given a Directed Graph with V vertices (Numbered from 0 to V-1) and E edges, check whether it contains any cycle or not.

### Example 1:
#### Input:

#### Output: 
1
#### Explanation: 
3 -> 3 is a cycle

### Example 2:

#### Input:
#### Output: 
0
#### Explanation: 
no cycle in the graph

### Your task:
You dont need to read input or print anything. Your task is to complete the function isCyclic() which takes the integer V denoting the number of vertices and adjacency list as input parameters and returns a boolean value denoting if the given directed graph contains a cycle or not.

**Expected Time Complexity:** O(V + E)

**Expected Auxiliary Space:** O(V)

### Constraints:
1 ≤ V, E ≤ 10^5

```python
# User function Template for python3
class Solution:
    
    def __init__(self):
        self.visited = None
        self.nodes_in_path = None
    
    
    def check_cycle_in_path(self, adj, node):
        
        # Mark the node as visited and add it in path
        self.visited[node] = True
        self.nodes_in_path[node] = True
        
        # For each neighboring node
        for neighbor in adj[node]:
            
            # Check previous path for visited node
            if self.visited[neighbor]:
                if self.nodes_in_path[neighbor]:
                    return True

            # Check the path ahead for unvisited node
            else:
                if self.check_cycle_in_path(adj, neighbor):
                    return True
    
        # Remove the node from path
        self.nodes_in_path[node] = False
                
        return False
    
    
    # Function to detect cycle in a directed graph.
    def isCyclic(self, V, adj):
        # code here
        self.visited = [False for i in range(V)]
        self.nodes_in_path = [False for i in range(V)]
        
        # For each component
        for i in range(V):
            
            # Check if the node is unvisited
            if not self.visited[i]:
                
                # Check for cycle in path ahead of current node
                if self.check_cycle_in_path(adj, i):
                    return True
                    
        return False

# { 
# Driver Code Starts
# Initial Template for Python 3
import sys
sys.setrecursionlimit(10**6)
if __name__ == '__main__':
    t = int(input())
    for i in range(t):
        V,E = list(map(int, input().strip().split()))
        adj = [[] for i in range(V)]
        for i in range(E):
            a,b = map(int,input().strip().split())
            adj[a].append(b)
        ob = Solution()
        if ob.isCyclic(V, adj):
            print(1)
        else:
            print(0)
# } Driver Code Ends
```
