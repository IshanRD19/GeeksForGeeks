# PRINT THIS PATTERN!

Given an integer N, you need to print the following pattern.

### Example 1:

#### Input:
N = 3
#### Output:
33333

32223

32123

32223

33333

#### Explanation:
When N = 3 then there will be three boundaries. Outer boundary contains only 3. Middle boundary contains only 2. And the inner boundary contains only 1.
### Example 2:

#### Input:
N = 4
#### Output:
4444444

4333334

4322234

4321234

4322234

4333334

4444444
### Your Task:  
You don't need to read input. Your task is to complete the function printPattern() which takes an integer N as an input parameter and print the pattern for it.

**Expected Time Complexity:** O(N^2)

**Expected Auxiliary Space:** O(1)

### Constraints:
1 <= N <= 9

```python
# User function Template for python3
class Solution:
    def printPattern(self, N):
        # code here
        total_chars = 2 * N - 1
        top_row = []

        for _ in range(total_chars):
            top_row.append(N)

        all_rows = [top_row]
        
        for i in range(1, N):
            new_row = all_rows[-1].copy()
            
            for j in range(i, N):
                new_row[j] -= 1
                
            row_mirror = new_row[:N-1]
            row_mirror.reverse()
            new_row[N:] = row_mirror

            all_rows.append(new_row)
            
        column_mirror = all_rows[:N-1]
        column_mirror.reverse()
        all_rows.extend(column_mirror)
        
        for row in all_rows:
            print(''.join([str(i) for i in row]))
            
        return


#{ 
 # Driver Code Starts
#Initial Template for Python 3

if __name__ == '__main__':
    T=int(input())
    for i in range(T):
        N = int(input())
        ob = Solution()
        ob.printPattern(N)

# } Driver Code Ends
```
