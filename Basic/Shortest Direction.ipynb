# SHORTEST DIRECTION

A person wants to go from origin to a particular location, he can move in only 4 directions(i.e East, West, North, South) but his friend gave him a long route, help a person to find minimum Moves so that he can reach to the destination.

**Note:** You need to print the lexicographically sorted string. Assume the string will have only ‘E’ ‘N’ ‘S’ ‘W’ characters.

### Example 1:

#### Input:
S = "SSSNEEEW"
#### Output: 
EESS
#### Explanation: 
Following the path SSSNEEEW
and EESS gets you at the same final point.
There's no shorter path possible.
### Example 2:

#### Input: 
S = "NESNWES"
#### Output: 
E
#### Explanation: 
Following the path NESNWES
and E gets you at the same final point.
There's no shorter path possible.

### Your Task:
You don't need to read input or print anything. Your task is to complete the function shortestPath() which takes the string S as input and returns the resultant string denoting the shortest path in lexicographic order.

**Expected Time Complexity:** O(|S|).

**Expected Auxiliary Space:** O(|S|) for output.

### Constraints:
1<=|S|<=10^5

```python
# User function Template for python3
class Solution:
    def shortestPath(self, s):
        # code here
        result = ''
        direction_count = {'N': 0, 'S': 0, 'E': 0, 'W': 0}
        meaningless = {'N': 'S', 'S': 'N', 'E': 'W', 'W': 'E'}

        for i in range(len(s)):
            direction_count[s[i]] += 1
            
        if direction_count['E'] > direction_count['W']:
            result += 'E' * (direction_count['E'] - direction_count['W'])
        
        if direction_count['N'] > direction_count['S']:
            result += 'N' * (direction_count['N'] - direction_count['S'])

        if direction_count['S'] > direction_count['N']:
            result += 'S' * (direction_count['S'] - direction_count['N'])

        if direction_count['W'] > direction_count['E']:
            result += 'W' * (direction_count['W'] - direction_count['E'])

        return result

# { 
# Driver Code Starts
# Initial Template for Python 3
if __name__ == '__main__':
    t = int(input())
    for _ in range(t):
        s = input().strip()
        ob = Solution()
        ans = ob.shortestPath(s)
        print(ans)
# } Driver Code Ends
```

```python
class Solution:
    
    def __init__(self):
        self.current_cherries = 0
        self.current_path = []
        self.max_cherries = 0
        self.max_cherry_path = 0
        

    def dfs(self, n, grid, cell, destination):
        
        # Check for boundary cells
        if not ((0 <= cell[0] < n) or (0 <= cell[1] < n)):
            return 0
        
        # Check for thorns
        if grid[cell[0]][cell[1]] == -1:
            return 0
        
        # Add cell in current path
        self.current_path.append(cell)
        self.current_cherries += grid[cell[0]][cell[1]]
        
        print(self.current_path, self.current_cherries)
        
        # Check for destination
        if cell == destination:
            
            # Compare cherries picked in current path
            if self.current_cherries > self.max_cherries:
                self.max_cherries = self.current_cherries
                self.max_cherry_path = self.current_path
                print('updated', self.max_cherry_path)

            
        else:

            # Go down
            if cell[0] + 1 < n:
                # self.current_cherries += self.dfs(n, grid, [cell[0]+1, cell[1]], destination)
                self.dfs(n, grid, [cell[0]+1, cell[1]], destination)

            # Go right
            if cell[1] + 1 < n:
                # self.current_cherries += self.dfs(n, grid, [cell[0], cell[1] + 1], destination)
                self.dfs(n, grid, [cell[0], cell[1] + 1], destination)
        
        # Remove current cell from path
        self.current_path.pop()
        self.current_cherries -= grid[cell[0]][cell[1]]
        
        print(self.current_path, self.current_cherries)

        return

    
    def dfs_return(self, n, grid, cell, destination):
        
        # Check for boundary cells
        if not ((0 <= cell[0] < n) or (0 <= cell[1] < n)):
            return 0
        
        # Check for thorns
        if grid[cell[0]][cell[1]] == -1:
            return 0
        
        # Add cell in current path
        self.current_path.append(cell)
        self.current_cherries += grid[cell[0]][cell[1]]
        
        print(self.current_path, self.current_cherries)
        
        # Check for destination
        if cell == destination:
            
            # Compare cherries picked in current path
            if self.current_cherries > self.max_cherries:
                self.max_cherries = self.current_cherries
                self.max_cherry_path = self.current_path
                print('updated', self.max_cherry_path)
            
        else:

            # Go up
            if cell[0] - 1 >= 0:
                self.dfs_return(n, grid, [cell[0] - 1, cell[1]], destination)

            # Go right
            if cell[1] - 1 >= 0:
                self.dfs_return(n, grid, [cell[0], cell[1] - 1], destination)
        
        # Remove current cell from path
        self.current_path.pop()
        self.current_cherries -= grid[cell[0]][cell[1]]
        
        print(self.current_path, self.current_cherries)

        return

    
    def cherryPickup(self, grid: List[List[int]]) -> int:
        n = len(grid)

        start_cell, destination = [0,0], [n-1, n-1]
        self.dfs(n, grid, start_cell, destination)
        from_result = self.max_cherries
        print('from_result', from_result)
        
        print('max p', self.max_cherry_path)
        
        for cell in self.max_cherry_path:
            grid[cell[0]][cell[1]] = 0
            
        print('new_grid', grid)
        
        self.current_cherries = 0
        self.current_path = []
        self.max_cherries = 0
        self.max_cherry_path = 0

        start_cell, destination = [n-1, n-1], [0,0]
        self.dfs_return(n, grid, start_cell, destination)
        to_result = self.max_cherries
        print('to_result', to_result)


        return from_result + to_result
```
