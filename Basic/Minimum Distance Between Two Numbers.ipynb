# MINIMUM DISTANCE BETWEEN TWO NUMBERS

You are given an array A, of N elements. Find minimum index based distance between two elements of the array, x and y.

### Example 1:

#### Input:
N = 4

A[] = {1,2,3,2}

x = 1, y = 2
#### Output:
1
#### Explanation:
x = 1 and y = 2. There are two distances between x and y, which are 1 and 3 out of which the least is 1.

#### Input:

x = 42, y = 12
#### Output:
-1
#### Explanation:
