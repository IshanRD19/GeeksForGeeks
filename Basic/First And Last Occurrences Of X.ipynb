# FIRST AND LAST OCCURRENCES OF X

Given a sorted array arr containing n elements with possibly duplicate elements, the task is to find indexes of first and last occurrences of an element x in the given array.

### Example 1:

#### Input:
n=9, x=5

arr[] = { 1, 3, 5, 5, 5, 5, 67, 123, 125 }
#### Output:
2 5
#### Explanation:
First occurrence of 5 is at index 2 and last occurrence of 5 is at index 5. 
 

### Example 2:

#### Input:
n=9, x=7

arr[] = { 1, 3, 5, 5, 5, 5, 7, 123, 125 }
#### Output:
6 6 

### Your Task:
Since, this is a function problem. You don't need to take any input, as it is already accomplished by the driver code. You just need to complete the function find() that takes array arr, integer n and integer x as parameters and returns the required answer.
Note: If the number x is not found in the array just return both index as -1.

**Expected Time Complexity:** O(logN)

**Expected Auxiliary Space:** O(1).

### Constraints:
1 ≤ N ≤ 10^7

```python
# User function Template for python3
def find(arr,n,x):
    # code here
    left, right = 0, n-1
    first = last = -1

    # Check if first element is first occurrence of x
    if arr[left] == x:
        first = left

    else:
        # Find the first occurrence of x
        while left <= right:
            middle = (left + right) // 2

            # Check if middle element is less than x
            if arr[middle] < x:
                left = middle + 1

            # Check if middle element is greater than x
            elif arr[middle] > x:
                right = middle - 1

            # Check if middle element is x
            else:
                first = min(first, middle) if first != -1 else middle

                if arr[middle-1] != x:
                    break
                else:
                    right = middle - 1

    # Prepare range for searching the last occurrence of x
    left, right = first, n-1

    # Check if last element is last occurrence of x
    if arr[right] == x:
        last = right
        
    else:
        # Find the last occurrence of x
        while left <= right:
            middle = (left + right) // 2

            # Check if middle element is less than x
            if arr[middle] < x:
                left = middle + 1

            # Check if middle element is greater than x
            elif arr[middle] > x:
                right = middle - 1

            # Check if middle element is x
            else:
                last = max(last, middle)

                if arr[middle+1] != x:
                    break
                else:
                    left = middle + 1
                    
    return [first, last]

#{ 
 # Driver Code Starts
t=int(input())
for _ in range(0,t):
    l=list(map(int,input().split()))
    n=l[0]
    x=l[1]
    arr=list(map(int,input().split()))
    ans=find(arr,n,x)
    print(*ans)
# } Driver Code Ends
```
