# PATTERN FOR REVERSE CHARACTER BRIDGE

For a given value N, denoting the number of Characters starting from the A, print reverse character bridge pattern.

### Example 1:

#### Input:
N = 5
#### Output:
ABCDEDCBA

ABCD DCBA

ABC   CBA

AB     BA

A       A
### Example 2:

#### Input:
N = 4
#### Output:
ABCDCBA

ABC CBA

AB   BA

A     A
 

### Your Task:
You don't need to read input or print anything. Your task is to complete the function revCharBridge() which takes an Integer N as input and returns a vector of strings where each line represents lines of the pattern. For example, if N = 3, the vector v = {"ABCBA", "AB BA", "A   A"}.

**Expected Time Complexity:** O(N)

**Expected Auxiliary Space:** O(N)

### Constraints:
2 <= N <= 26

```python
# User function Template for python3
class Solution:
    def revCharBridge(self, N):
        # code here 
        result = []

        for i in range(0, N):
            word = ''
            spaces = ''
            reverse = ''
            
            for j in range(N - i):
                word += chr(65 + j)
                
            if i:
                spaces += ' ' * (2 * i - 1)
                reverse = word[::-1]
            else:
                reverse = word[:-1][::-1]

            word += spaces
            word += reverse
            
            result.append(word)

        return result

#{ 
#  Driver Code Starts
#Initial Template for Python 3

if __name__ == '__main__': 
    t = int (input ())
    for _ in range (t):
        N=int(input())
        
        ob = Solution()
        ptr = ob.revCharBridge(N)
        for i in ptr:
            print(i)
# } Driver Code Ends
```
